```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import scipy.optimize

%matplotlib inline
```

```python
data_poland = pd.read_csv('corona.csv')

date = np.array(data_poland['date'])
deaths = np.array(data_poland['n'])
```

```python
def func_expo(x, a, b):
    return a * np.exp(b * x)

def func_logistic(x, x0, k, V):
    return V/(1+np.exp(k*(x-x0)))
```

```python
scipy.optimize.curve_fit(func_expo,  date,  deaths)
```

```python
scipy.optimize.curve_fit(func_logistic,  date,  deaths)
```

```python
a = 12.49564488
b = 0.17542295
x = np.linspace(0,30,101)
y_expo = func_expo(x, a, b)


x0 = 2.77333508e+01 
k = -2.40558725e-01
V = 3.11103085e+03
x = np.linspace(0,30,101)
y_logistic = func_logistic(x,x0,k,V)



fig, ax = plt.subplots(ncols=2, figsize = (15,4))
ax[0].scatter(date, deaths, label="data points", color = "red")
ax[0].plot(x, y_expo, label="exponential fit", color = 'gray')
ax[0].legend()

ax[1].scatter(date, deaths, label="data points", color = "red")
ax[1].plot(x, y_logistic, label="logistic function fit", color = 'gray')
ax[1].legend()

plt.show()
```
