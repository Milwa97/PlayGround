```python
import numpy as np
import matplotlib.pyplot as plt


from  matplotlib.colors import LinearSegmentedColormap

cmap = LinearSegmentedColormap.from_list('myCMAP1',['gray', 'pink'], N = 2)

np.random.seed(17)
```

```python
size_x = 50
size_y = 50
N = size_x * size_y
```

```python
## global Ising Lattice

IsingLattice = np.random.randint(0,2, size=(size_x+1, size_y+1) ) *2-1   
```

```python
def show_lattice():      
    fig, ax = plt.subplots()
    ax.imshow(IsingLattice, cmap=cmap, origin = 'lower')        
    ax.axis('off')
    fig.tight_layout()    
     
    
def update_lattice(x,y, new_value):
    IsingLattice[x][y] = new_value
    
    
             
def get_neighbors(x,y):    
    return [(x, (y+1) % size_y ), 
            (x, (y-1) % size_y ), 
            ((x+1) % size_x, y), 
            ((x-1) % size_x, y)]
                 
```

```python
show_lattice()
```

```python
class Spin():
    def __init__(self, x, y, state, neighbors, couplings): 
        
        self.x = x
        self.y = y
        self.state = state
        self.neighbors = neighbors
        
        self.couplings =couplings
        self.neighbors_state = np.zeros_like(couplings)

    
    def update_neighbors_state(self):
        for i, neighbor in enumerate(self.neighbors):
            x,y = neighbor
            self.neighbors_state[i] = IsingLattice[x][y]
    
    
    def update_spin_state(self):
         self.state *= -1
        

    def get_spin_state(self):
        return self.state
    
    
    def get_spin_neighbors(self):
        return self.neighbors

    
    def get_spin_energy(self):
        ''' H = -Jij si sj'''       
        energy = -np.dot(self.couplings, self.state * self.neighbors_state)
        print(energy)
        return energy  # -np.dot(self.couplings, self.state * self.neighbors_state)
    
    def __repr__(self):
        return repr((self.x, self.y, self.state))

    
    
```

```python
spins = []

for x in range(size_x):
    for y in range(size_y):
        neighbors = get_neighbors(x,y)
        couplings = np.ones(4)*1.0
        state = IsingLattice[x][y]            
        spins.append(Spin(x,y, state, neighbors,couplings))           
```

```python
for k in range(1000):
    show_lattice()
    

    index = np.random.randint(N)
    spin = spins[index]
    spin.update_neighbors_state()
    
    if spin.get_spin_energy() > 0:
        spin.update_spin_state()
        update_lattice(spin.x, spin.y, spin.state)

```
