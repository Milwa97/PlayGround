```python
import numpy as np
import matplotlib.pyplot as plt
```

## Bifurcation diagram:

$$x_{n+1} = r\cdot x_n(1-x_n)$$

$r\in (0,4)$

$x_n \in (0,1)$

```python
x0 = 0.7

r_min = 0
r_max = 4.0
m = 400
r_tab = np.linspace(r_min, r_max, m)

delay = 50  ## time delay
p = 200   ## number of points
```

```python
%%time
fig, ax = plt.subplots(figsize = (18,8))
ax.set_xlabel('r')
ax.set_ylabel('x')
ax.set_xlim(-0.1, 4.1)
ax.set_ylim(-0.1, 1.1)

for r in r_tab:
    x = x0
    
    if r < 2.9:
        p = 5   ## number of points
    else: 
        p = 200
        
    for i in range(p + delay):
        x = r*x*(1-x)
        if i > delay:
            ax.scatter(r, x, s = 0.5, c='deeppink', alpha = 0.8)   
plt.show()
```

## Bifurcation diagram:

$$x_{n+1} = \frac{\mu}{2}\cdot \left( 1- 2\cdot |x_n-\frac{1}{2}|) \right)$$

$\mu \in (0,2)$

$x_n \in (0,1)$

```python
x0 = 0.65

mu_min = 0
mu_max = 2.0
m = 400
mu_tab = np.linspace(mu_min, mu_max, m)

delay = 50  ## time delay
p = 200   ## number of points
```

```python
%%time
fig, ax = plt.subplots(figsize = (18,8))
ax.set_xlabel('$\mu$')
ax.set_ylabel('x')
ax.set_xlim(-0.1, 2.1)
ax.set_ylim(-0.1, 1.1)

for mu in mu_tab:
    x = x0
    
    for i in range(p + delay):
        x = mu * 0.5 * (1-2*abs(x-0.5))
        if i > delay:
            ax.scatter(mu, x, s = 0.5, c = 'purple', alpha = 0.8)   
plt.show()
```

## Bifurcation diagram:

$$x_{n+1} = r \left( 1-\left( 2x-1 \right)^4  \right)$$

$r \in (0,1)$

$x_n \in (0,1)$

```python
x0 = 0.7

r_min = 0
r_max = 1.0
m = 400
r_tab = np.linspace(r_min, r_max, m)

delay = 50  ## time delay
p = 200   ## number of points
```

```python
%%time
fig, ax = plt.subplots(figsize = (18,8))
ax.set_xlabel('r')
ax.set_ylabel('x')
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)

for r in r_tab:
    x = x0
    for i in range(p + delay):
        x = r*(1- (2*x-1)**4)
        if i > delay:
            ax.scatter(r, x, s = 0.5, c='green', alpha = 0.8)   
plt.show()
```

## Bifurcation diagram:

$$x_{n+1} = \frac{r}{4} \sin(\pi x)$$

$r \in (0,4)$

$x_n \in (0,1)$

```python
x0 = 0.7

r_min = 0
r_max = 4.0
m = 400
r_tab = np.linspace(r_min, r_max, m)

delay = 50  ## time delay
p = 200   ## number of points
```

```python
%%time
fig, ax = plt.subplots(figsize = (18,8))
ax.set_xlabel('r')
ax.set_ylabel('x')
ax.set_xlim(-0.1, 4.1)
ax.set_ylim(-0.1, 1.1)

for r in r_tab:
    x = x0
    
    for i in range(p + delay):
        x = r*0.25*np.sin(np.pi * x)
        if i > delay:
            ax.scatter(r, x, s = 0.5, c='greenyellow', alpha = 0.8)   
plt.show()
```
